```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/content/drive/MyDrive/Datasets/diabetes.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.describe()
```

```python
scaler = MinMaxScaler()
model = scaler.fit(data)
scaled_data = model.transform(data)
```

```python
normalized_data = pd.DataFrame(scaled_data, columns=data.columns)
```

```python
normalized_data.head()
```

```python
normalized_data.describe()
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
sns.set_theme()
sns.histplot(ax=ax[0], x=data['BloodPressure'], kde=True)
ax[0].set_title('Regular Data')

sns.histplot(ax=ax[1], x=normalized_data['BloodPressure'], kde=True)
ax[1].set_title('Normalized Data')

plt.tight_layout()
plt.show()
```
