# **MNIST Single digit data visualization**

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset from google drive
mnist_data = pd.read_csv('/content/drive/MyDrive/Datasets/mnist_test.csv')
mnist_data.shape
```

MNIST test Dataset contains :
  - 10000 Rows
  - 785 Columns

```python
mnist_data.head()
```

```python
row = mnist_data.iloc[4] # Extract the selected row

label = row['label'] # Separate the label and pixel values

pixel_values = row[1:].values # Extract all pixel columns

image_array = pixel_values.reshape(28,28) # Reshape pixel values into a 28x28 array
```

```python
# Data Visualizing using seaborn's heatmap

plt.figure(figsize=(6,6))
sns.heatmap(image_array, cmap='gray',cbar=False, square=True, xticklabels=False, yticklabels=False)
plt.title(f'Digit: {label}')
plt.show()
```
